```python
#1. What is hypothesis testing in statistics?
""" Hypothesis testing in statistics is a method used to make decisions about a population based on sample data. It involves:
- Formulating a null hypothesis (H0) and an alternative hypothesis (H1)
- Collecting sample data
- Calculating a test statistic and p-value
- Deciding whether to reject the null hypothesis based on the p-value and a chosen significance level
This helps determine if there's enough evidence to support a claim about the population."""
```

```python
#2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
"""In hypothesis testing:
- Null Hypothesis (H0): A statement of no effect or no difference. It's what you aim to test against .
- Alternative Hypothesis (H1): A statement that contradicts the null hypothesis, suggesting there is an effect or difference .
For example, if testing a new medicine:
- H0: The medicine has no effect.
- H1: The medicine has an effect.
You try to find evidence against H0 to support H1.
Or in simple words:-
Null Hypothesis (H0):
1. Statement of no effect or no difference
2. What you aim to test against
3. Typically states equality (e.g., =, ≤, ≥)
Alternative Hypothesis (H1):
1. Statement that contradicts the null hypothesis
2. Suggests there is an effect or difference
3. Typically states inequality (e.g., ≠, <, >)
Example:
H0: μ = 50 (no difference)
H1: μ ≠ 50 (difference exists)"""
```

```python
#3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
""" The significance level, often denoted as α (alpha), is a threshold probability value used in hypothesis testing to determine whether the null hypothesis should be rejected.
Role of Significance Level:
1. Decision criterion: α determines the maximum probability of rejecting the null hypothesis when it is true (Type I error).
2. Test sensitivity: A smaller α (e.g., 0.01) requires stronger evidence to reject H0, while a larger α (e.g., 0.10) requires less evidence.
Common significance levels:
1. 0.05 (5%): Most commonly used threshold
2. 0.01 (1%): More stringent threshold
3. 0.10 (10%): Less stringent threshold
Outcome decision:
1. If p-value ≤ α, reject H0 (statistically significant result)
2. If p-value > α, fail to reject H0 (no statistically significant result)
The significance level helps balance the risk of Type I errors (false positives) and the ability to detect true effects."""
```

```python
#4. What are Type I and Type II errors? Give examples of each.
""" Type I Error:
1. Rejecting a true null hypothesis (False Positive)
2. Example:
- Null Hypothesis (H0): A person is innocent.
- Alternative Hypothesis (H1): A person is guilty.
- Type I Error: Convicting an innocent person.
Type II Error:
1. Failing to reject a false null hypothesis (False Negative)
2. Example:
- Null Hypothesis (H0): A person is not infected with a disease.
- Alternative Hypothesis (H1): A person is infected with a disease.
- Type II Error: Failing to diagnose a person who actually has the disease.
These errors have significant implications in various fields, such as medicine, law, and quality control."""
```

```python
#5. What is the difference between a Z-test and a T-test? Explain when to use each.
""" Z-test vs T-test:
Z-test:
1. Used when:
- Population variance is known
- Sample size is large (n > 30)
- Data follows a normal distribution
2. Compares sample mean to population mean or two sample means
T-test:
1. Used when:
- Population variance is unknown
- Sample size is small (n < 30)
- Data follows a normal distribution or approximately normal
2. Compares sample mean to population mean, two sample means, or paired samples
Key difference:
The Z-test uses the standard normal distribution (Z-distribution), while the T-test uses the T-distribution, which accounts for the added uncertainty due to estimating population variance from the sample.
When to use each:
1. Z-test: Large samples, known population variance, or specific population comparisons.
2. T-test: Small samples, unknown population variance, or paired comparisons.
Choose the test based on your sample size, data distribution, and whether the population variance is known."""
```

```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
#(Include your Python code and output in the code box below.)
#Hint: Generate random number using random function.
import numpy as np
import matplotlib.pyplot as plt
# Set seed for reproducibility
np.random.seed(0)
# Parameters
n = 10  # number of trials
p = 0.5  # probability of success
size = 1000  # number of experiments
# Generate binomial distribution
binomial_data = np.random.binomial(n, p, size)
# Plot histogram
plt.hist(binomial_data, bins=11, edgecolor='black', align='left', rwidth=0.8)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
#7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
#(Include your Python code and output in the code box below.)
import numpy as np
from scipy import stats
# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Null and alternative hypotheses
# H0: μ = 50 (population mean is 50)
# H1: μ ≠ 50 (population mean is not 50)
# Population standard deviation (known)
population_std_dev = 1
# Sample size
n = len(sample_data)
# Sample mean
sample_mean = np.mean(sample_data)
# Calculate Z-statistic
z_statistic = (sample_mean - 50) / (population_std_dev / np.sqrt(n))
# Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
# Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (H0). The sample mean is significantly different from 50.")
else:
    print("Fail to reject the null hypothesis (H0). The sample mean is not significantly different from 50.")
```

```python
#8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#(Include your Python code and output in the code box below.)
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
# Set seed for reproducibility
np.random.seed(0)
# Parameters
mu = 5  # population mean
sigma = 2  # population standard deviation
n = 100  # sample size
# Generate data from normal distribution
data = np.random.normal(mu, sigma, n)
# Calculate sample mean and standard error
sample_mean = np.mean(data)
standard_error = sigma / np.sqrt(n)
# Calculate 95% confidence interval
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)
# Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"95% Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")
# Plot histogram
plt.hist(data, bins=20, edgecolor='black', alpha=0.7, label='Data')
plt.axvline(sample_mean, color='red', label='Sample Mean')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label='95% CI')
plt.axvline(confidence_interval[1], color='green', linestyle='--')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data with 95% Confidence Interval')
plt.show()
```

```python
#9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram.
#Explain what the Z-scores representin terms of standard deviations from the mean.
#(Include your Python code and output in the code box below.)
import numpy as np
import matplotlib.pyplot as plt
# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores
# Generate sample data
np.random.seed(0)
data = np.random.normal(5, 2, 1000)
# Calculate Z-scores
z_scores = calculate_z_scores(data)
# Print statistics
print(f"Mean of Z-scores: {np.mean(z_scores):.4f}")
print(f"Standard Deviation of Z-scores: {np.std(z_scores):.4f}")
# Plot histogram of original data
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Plot histogram of Z-scores
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
plt.title('Standardized Data (Z-scores)')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.show()

#What Z-scores represent:
"""Z-scores represent the number of standard deviations a data point is away from the mean. A Z-score:
- of 0 means the data point is equal to the mean.
- greater than 0 means the data point is above the mean.
- less than 0 means the data point is below the mean.
- of 1 or -1 means the data point is 1 standard deviation away from the mean.
- of 2 or -2 means the data point is 2 standard deviations away from the mean, and so on.
By standardizing data using Z-scores, we can compare data points from different distributions or datasets on the same scale."""
```
